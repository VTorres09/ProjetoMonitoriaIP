### Como foi solicitado POO, criei classes com os atributos sendo as colunas do DataFrame e objetos sendo as linhas.

```python
#Criando a classe carro para validacao
class Carro():

    def __init__(self,ano,lugar,cor,km,modelo,preco):

        self.ano = ano
        self.lugar = lugar
        self.cor = cor
        self.km = km
        self.modelo = modelo
        self.preco = preco
    
#Criando metodos   
    def getMenorAno(self): #Utilizado na criação do último gráfico
        x = self.ano 
        y = x.split('/')
        z = y[0].split(' ')
        return z[0]
    def getLugar(self): 
        return self.lugar
    def getCor(self): 
        return self.cor
    def getKm(self): 
        return self.km
    def getModelo(self): 
        return self.modelo
    def getPreco(self): 
        return self.preco

```

### Plotando um gráfico baseado na quantidade de cores dos carros

```python
pylab.rcParams['figure.figsize'] = (6.0, 6.0)
plt.style.use("fivethirtyeight")
plt.style.use("ggplot")
explode = (0.1, 0, 0, 0)
concatenado['Cor'].value_counts().plot.pie(wedgeprops = dict(width=0.7), autopct='%1.1f%%',
        shadow=True, startangle=90, explode=explode)
plt.title('Cores dos Carros')
```

### Plotando um gráfico com base no preço dos automóveis.

Obs: Foi necessária a conversão do tipo da coluna para ordenar o eixo X no gráfico.

```python
#Deletando colunas que nao possuem dados de preco.
concatenado2 = concatenado.drop([5,6,7,8,9,10,11])
Index = [0,1,2,3,4,5,6,7]
concatenado2.insert(loc=0, column = 'Index', value = Index)
concatenado2 = concatenado2.set_index('Index')

#Removendo os 'R$' e ',' da coluna 'Preco' para transformar no tipo Int.
for i in range(0,8):
    x = concatenado2.loc[i, 'Preco'].split(' ')
    y = x[1].split(',')
    z = y[0].split('.')
    concatenado2.loc[i, 'Preco'] = z[0]+z[1]
    
#Trasformando a coluna 'Preco' para o tipo Int
concatenado2['Preco'] = concatenado2['Preco'].astype(int)

#Definindo variaveis
y = concatenado2['Preco']
x = concatenado2['Modelo']

#Criando grafico
plt.rcParams['axes.facecolor'] = 'black'
plt.rcParams['figure.facecolor'] = 'white'
plt.xlabel('Preco')
plt.ylabel('Modelo')
plt.title('Modelo de automoveis e seus respectivos precos.')
plt.barh(x,y)
```
